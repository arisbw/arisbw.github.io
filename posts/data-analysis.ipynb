# Basic Data Analysis with Python

In this post, we'll explore some basic data analysis using Python. We'll use popular libraries like pandas and matplotlib to analyze and visualize data.

## Setup

First, let's import the required libraries and create some sample data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample data
data = {
    'year': [2020, 2021, 2022, 2023, 2024],
    'sales': [100, 120, 140, 180, 210]
}

df = pd.DataFrame(data)
print("Our sample data:")
df
```

## Data Visualization

Now let's create a line plot to visualize the sales growth over the years.

```python
# Create a simple line plot
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['sales'], marker='o')
plt.title('Sales Growth Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

## Conclusion

We can see a clear upward trend in sales from 2020 to 2024, with particularly strong growth in recent years.
